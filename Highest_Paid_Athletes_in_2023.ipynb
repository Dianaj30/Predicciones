```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
```

```python
filename = "/content/drive/MyDrive/CODING DOJO-BOOTCAMP DATA ANALYTICS/Ejercicios plataforma Coding/The_World's_Highest_Paid_Athletes.xlsx"
df = pd.read_excel(filename)
df.head()
```

# EXPLORACIÓN DATA

```python
df.info()
```

```python
df['Total Earnings'].value_counts()
```

```python
df.describe()
```

```python
df.set_index('Rank', inplace=True)
```

```python
df.head()
```

```python
df_group = df.groupby(['Sport'])[['Total Earnings']].mean().reset_index()
df_group
```

```python
df_group.sort_values(by=['Total Earnings'], ascending=False)
```

# VISUALIZACIÓN DATA

```python
counts=df['Sport'].value_counts()
counts
```

```python
counts = counts.sort_values(ascending=True)
counts
```

```python
plt.barh(counts.index, counts.values, color='green')
plt.xlabel('Count')
plt.ylabel('Sport')
plt.title('Count of Sports')
plt.tight_layout()
plt.show()
```

```python
df_group = df_group.sort_values(by='Total Earnings', ascending=True)
df_group
```

```python
plt.barh(df_group['Sport'], df_group['Total Earnings'], color='lightgreen')
plt.xlabel('Paid Mean')
plt.ylabel('Sport')
plt.title('Paid Mean by Sport')
plt.tight_layout()
plt.show()
```
